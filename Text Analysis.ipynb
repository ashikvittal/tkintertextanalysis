```python
'''
Title: Text Analysis Application

Program to create a text analysis application, that will accept a text file and provide the following output,
1.total # of words
2.total # of occurrences for each word.
3.total # of characters in the text document
4.total # of blank spaces in the text document.
5.o	total # of blank spaces divided by total # of characters, multiplied by 100.
6. The output can be saved in the selected path in the user's computer


Author: Ashik Vittal Poovappa
Date: 11/25/2018
'''



from tkinter import Button, Tk, HORIZONTAL
import tkinter as tk
import re
from tkinter import filedialog
from tkinter import *
import time
import threading
from tkinter import messagebox

#----------------------------------------------------------------------------------------------------------

class main_GUI():
    def __init__(self,master):
        self.master=master
        #master.geometry("350x350+120+120")
        master.title("Text Analysis")
        master.minsize(width=400, height=200)
        master.maxsize(width=450, height=450)
        scroll=Scrollbar(master)
               
            
        '''Select file Label'''
        self.label1=tk.Label(master,text="Please select your file",font=("Calibri", 10))
        self.label1.grid(row=0, column=0,sticky=W+E,padx=5, pady=5)
        
        '''Browse buttonw'''
        self.e=tk.Entry(master, text="").grid(row=1, column=0,sticky=N+S+E+W)
        
        self.button = tk.Button(master, text="Browse", command=self.openfile,  
                                background = 'white', foreground = "black")
        self.button.grid(row=1, column=1,sticky=N+S+E+W)
        
        '''MENU BAR'''
        main_menu=Menu(master)
        master.config(menu=main_menu)
        fileMenu=Menu(main_menu)
        
        main_menu.add_cascade(label="File", menu=fileMenu) #file menu
        main_menu.add_cascade(label="Save As")#saveas
        main_menu.add_cascade(label="Quit", command=root.destroy)#Destroy
        
        '''List box to display the results'''
        listFrame = Frame(master)
        listFrame.grid(row=6, column=0, sticky="ns")
        
        scroll= Scrollbar(listFrame, orient='vertical')
        scroll.grid(row=6, column=1,sticky="ns")
        
        
        self.fileList = Listbox(listFrame, width=60, height=10, yscrollcommand=scroll.set)
        self.fileList.grid(row=6, column=0, sticky="ns",padx=10, pady=10)
        scroll.config(command=self.fileList.yview)
        
        
        
    def openfile(self):
        try:
            path=filedialog.askopenfilename()
        except "FileNotFoundError":    #file not found error handelling
            self.fileList.insert(END,"File Not Found")
            
        file_name=path.split("/")[-1]
        self.fileName=file_name.split(".")[-2]
        
        self.fileNameLabel=tk.Label(text="Selected File: {}".format(self.fileName),font=("Calibri", 10))
        self.fileNameLabel.grid(row=3, column=0,sticky=W+E,padx=10, pady=10)
        
        self.fileList.insert(END,"1. Selected File: {}".format(self.fileName))#listbox update 1
        
        '''confirmation to continue using message box'''
        answer = messagebox.askyesno("Continue","Click Yes to continue and No to select new file")
        if answer==True:
            self.fileList.insert(END,"2. Continue? :Yes") #listbox update 2
        elif answer==False:
            root.destroy()
        
        
        
        '''Reading the file'''
        try:  
            with open(path,"r") as file:
                data=file.read()
        except IOError:
            self.fileList.insert(END,"Error reading the file, retry or select a different file.")
                
        self.fileList.insert(END,"3. Reading Document") #listbox update 3
        f=data.replace("\n","")
        f=f.lower()
        self.charCount=len(f)

        '''removing all punctuations and symbols'''
        cleanFile = re.sub(r'[^\w\s]','',f)
        
        wordCount=cleanFile.split(" ")
        
        self.spaceCount=data.count(" ")   #count blank spaces
        '''counting occurances of each word'''
        self.wordCount={}
        for x in cleanFile.split():
            if x not in self.wordCount:
                self.wordCount[x] = 1
            else:
                self.wordCount[x] += 1
            
        
        '''percentage of blank spaces calculations'''
        
        try:
            self.perBlank=(self.spaceCount/self.charCount)*100  
        except ZeroDivisionError:  #Zero division error handelling
            self.perBlank=0
            
        self.words=len(wordCount)
        
        self.fileList.insert(END,"4. your result is ready")#listbox update 4
        
        '''Download button'''
        self.button1 = tk.Button(text="Download result", command =self.export,  
                                 background = 'white', foreground = "black") 
        self.button1.grid(row=5, column=0,sticky=W+E,padx=10, pady=10)
        
    '''Exporting the results'''
    def export(self):
        '''results is a dictionary containing all the analysis results'''
        self.results={"Total_Words":self.words,"Total_Characters":self.charCount,"Total_Blank_Spaces":self.spaceCount,
                         "Percentage_Of_Blank_Spaces":self.perBlank}
                 
                 
        file=filedialog.asksaveasfile(mode="w", defaultextension=".txt", title="Save Analysis", 
                                      initialfile='{}_analysis'.format(self.fileName)) #find the location to save
        
        if file is None: #if abandoned in the middle of the function
            return
        
        for key, value in self.results.items():
            file.write("{}: ".format(key))
            file.write("{:,}".format(value))

            
        file.write("\n\n Total Occurances of each word:\n")
        
        '''using the dicionary created in the openFile function to write the occurance of each word'''         
        for key,value in self.wordCount.items():
            file.write("{}:".format(key))
            file.write("{}\n\n".format(value))
                

    
        
root=tk.Tk()
app=main_GUI(root)
root.mainloop()
        
        
```
